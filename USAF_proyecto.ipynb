Introducción a la USAF

La Fuerza Aérea de los Estados Unidos (USAF, United States Air Force) es la rama aérea de las fuerzas armadas de los Estados Unidos. Fue establecida como una fuerza militar independiente el 18 de septiembre de 1947, tras la aprobación de la Ley de Seguridad Nacional de 1947, separándose del Ejército de los Estados Unidos.

La USAF es una de las fuerzas aéreas más grandes y tecnológicamente avanzadas del mundo. Su misión principal es defender el espacio aéreo estadounidense, apoyar operaciones militares en todo el mundo, realizar misiones de transporte estratégico, reconocimiento, ataque y disuasión nuclear.

Opera una gran variedad de aeronaves militares, entre las que se incluyen:

Cazas de superioridad aérea y combate (como F-22, F-35 y F-15).

Bombarderos estratégicos (como B-2, B-1 y B-52).

Aviones de transporte militar (como C-17 y C-130).

Aviones de reabastecimiento en vuelo (como KC-135 y KC-46).

Aeronaves de vigilancia y reconocimiento (como E-3 AWACS y RQ-4 Global Hawk).

Gracias a esta diversidad de aeronaves y a su alto nivel tecnológico, la USAF cumple un papel fundamental en la defensa, proyección de poder y operaciones aéreas de los Estados Unidos.

```python
# Definimos una pequeña 'base de conocimiento' sobre aviones de combate de la USAF
fighter_jets = {
    "F-15 Eagle": {
        "rol": "Caza de superioridad aérea",
        "caracteristica_clave": "Gran velocidad y maniobrabilidad, sin derribos en combate aéreo",
        "generacion": "Cuarta",
        "velocidad_maxima": "Mach 2.5+ (aprox. 2655 km/h)",
        "techo_operativo": "65,000 ft (aprox. 19,812 m)",
        "motores": "2 × Pratt & Whitney F100 (o General Electric F110)",
        "dato_historico": "Ha derribado más de 100 aviones enemigos en combate aéreo sin sufrir ninguna pérdida propia.",
        "estado_operativo": "Operativo"
    },
    "F-15EX Eagle II": {
        "rol": "Caza de superioridad aérea, ataque multi-rol",
        "caracteristica_clave": "Versión avanzada del F-15, con electrónica moderna y mayor carga de armas",
        "generacion": "Cuarta++",
        "velocidad_maxima": "Mach 2.5+ (aprox. 2655 km/h)",
        "techo_operativo": "65,000 ft (aprox. 19,812 m)",
        "motores": "2 × General Electric F110-GE-129",
        "dato_historico": "Diseñado para complementar a los cazas de quinta generación y permanecer en servicio hasta 2040 y más allá.",
        "estado_operativo": "Operativo"
    },
    "F-16 Fighting Falcon": {
        "rol": "Caza polivalente",
        "caracteristica_clave": "Versátil y ampliamente utilizado, coste efectivo",
        "generacion": "Cuarta",
        "velocidad_maxima": "Mach 2.05 (aprox. 2170 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "1 × Pratt & Whitney F100 o General Electric F110",
        "dato_historico": "Uno de los cazas más producidos y exportados del mundo, sirviendo en más de 25 fuerzas aéreas.",
        "estado_operativo": "Operativo"
    },
    "F-18 Super Hornet": {
        "rol": "Caza multi-rol embarcado",
        "caracteristica_clave": "Principal caza de la US Navy, polivalente, adaptable a portaaviones",
        "generacion": "Cuarta",
        "velocidad_maxima": "Mach 1.8 (aprox. 1915 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "2 × General Electric F414-GE-400",
        "dato_historico": "Es la columna vertebral de la aviación naval de EE. UU., famoso por su robustez en operaciones desde portaaviones.",
        "estado_operativo": "Operativo"
    },
    "F-22 Raptor": {
        "rol": "Caza de superioridad aérea, ataque a tierra",
        "caracteristica_clave": "Sigilo, supercrucero, alta maniobrabilidad",
        "generacion": "Quinta",
        "velocidad_maxima": "Mach 2.25+ (supercrucero Mach 1.82)",
        "techo_operativo": "65,000 ft (aprox. 19,812 m)",
        "motores": "2 × Pratt & Whitney F119-PW-100",
        "dato_historico": "Fue el primer caza de quinta generación en entrar en servicio y es considerado uno de los aviones de combate más dominantes jamás construidos.",
        "estado_operativo": "Operativo"
    },
    "F-35A Lightning II": {
        "rol": "Caza polivalente furtivo (CTOL)",
        "caracteristica_clave": "Sigilo, fusión de sensores, despegue y aterrizaje convencional",
        "generacion": "Quinta",
        "velocidad_maxima": "Mach 1.6 (aprox. 1930 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "1 × Pratt & Whitney F135-PW-100",
        "dato_historico": "Su diseño se enfoca en la fusión de sensores, permitiendo a los pilotos una conciencia situacional sin precedentes.",
        "estado_operativo": "Operativo"
    },
    "F-35B Lightning II": {
        "rol": "Caza polivalente furtivo (STOVL)",
        "caracteristica_clave": "Sigilo, fusión de sensores, despegue corto y aterrizaje vertical",
        "generacion": "Quinta",
        "velocidad_maxima": "Mach 1.6 (aprox. 1930 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "1 × Pratt & Whitney F135-PW-600 con sistema de elevación de ventilador",
        "dato_historico": "Es el único caza supersónico operativo del mundo con capacidad STOVL.",
        "estado_operativo": "Operativo"
    },
    "F-35C Lightning II": {
        "rol": "Caza polivalente furtivo (CATOBAR)",
        "caracteristica_clave": "Sigilo, fusión de sensores, adaptable a portaaviones con catapulta",
        "generacion": "Quinta",
        "velocidad_maxima": "Mach 1.6 (aprox. 1930 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "1 × Pratt & Whitney F135-PW-100",
        "dato_historico": "Diseñado específicamente para operaciones desde portaaviones grandes con alas más grandes y tren de aterrizaje reforzado.",
        "estado_operativo": "Operativo"
    },
    "F-117 Nighthawk": {
        "rol": "Avión de ataque furtivo",
        "caracteristica_clave": "El primer avión operativo diseñado con sigilo, forma distintiva",
        "generacion": "Tercera/Cuarta (diseño especializado de sigilo)",
        "velocidad_maxima": "Mach 0.92 (aprox. 1100 km/h)",
        "techo_operativo": "45,000 ft (aprox. 13,716 m)",
        "motores": "2 × General Electric F404-GE-F1D2",
        "dato_historico": "Fue el primer avión operativo diseñado completamente alrededor de la tecnología furtiva, ganándose el apodo de \"Avión Invisible\".",
        "estado_operativo": "Retirado"
    },
    "A-10 Warthog": {
        "rol": "Apoyo aéreo cercano",
        "caracteristica_clave": "Armado con cañón GAU-8 Avenger, muy resistente",
        "generacion": "Tercera/Cuarta (diseño especializado)",
        "velocidad_maxima": "706 km/h (a nivel del mar)",
        "techo_operativo": "45,000 ft (aprox. 13,716 m)",
        "motores": "2 × General Electric TF34-GE-100",
        "dato_historico": "Construido alrededor de su cañón GAU-8 Avenger de 30 mm, es un tanque volador blindado diseñado para resistir grandes daños.",
        "estado_operativo": "Operativo"
    },
    "AV-8B Harrier II": {
        "rol": "Avión de ataque ligero/reconocimiento (STOVL)",
        "caracteristica_clave": "Capacidad de despegue corto y aterrizaje vertical/rodado, usado por Marines",
        "generacion": "Segunda/Tercera",
        "velocidad_maxima": "Mach 0.9 (aprox. 1083 km/h)",
        "techo_operativo": "49,000 ft (aprox. 14,935 m)",
        "motores": "1 × Rolls-Royce F402-RR-408 (Pegasus)",
        "dato_historico": "Famoso por su capacidad de despegue y aterrizaje vertical/corto, lo que le permite operar desde una variedad de lugares.",
        "estado_operativo": "Operativo (con Marines de EE. UU.)"
    },
    "YF-23 Black Widow II": {
        "rol": "Prototipo de caza de superioridad aérea",
        "caracteristica_clave": "Tecnología furtiva avanzada, alta velocidad, no entró en producción",
        "generacion": "Quinta (prototipo)",
        "velocidad_maxima": "Mach 2.2 (estimado)",
        "techo_operativo": "65,000 ft (aprox. 19,812 m)",
        "motores": "2 × General Electric YF120 o Pratt & Whitney YF119 (prototipos)",
        "dato_historico": "Aunque perdió el concurso ATF contra el F-22, muchos creen que era aún más furtivo y rápido.",
        "estado_operativo": "Prototipo"
    },
    "B-1B Lancer": {
        "rol": "Bombardero estratégico de largo alcance",
        "caracteristica_clave": "Bombardero supersónico, alas de geometría variable, gran carga útil convencional",
        "generacion": "Tercera/Cuarta",
        "velocidad_maxima": "Mach 1.25 (aprox. 1330 km/h)",
        "techo_operativo": "60,000 ft (aprox. 18,288 m)",
        "motores": "4 × General Electric F101-GE-102",
        "dato_historico": "Conocido como el \"Bone\", es el pilar de la flota de bombarderos de largo alcance de la USAF para municiones convencionales.",
        "estado_operativo": "Operativo"
    },
    "B-2 Spirit": {
        "rol": "Bombardero estratégico furtivo",
        "caracteristica_clave": "Diseño de ala volante, sigilo extremo, capacidad nuclear y convencional",
        "generacion": "Cuarta (diseño furtivo)",
        "velocidad_maxima": "Mach 0.95 (aprox. 1010 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "4 × General Electric F118-GE-100",
        "dato_historico": "Su diseño único de ala volante y su sigilo extremo lo hacen casi indetectable por radar.",
        "estado_operativo": "Operativo"
    },
    "B-52 Stratofortress": {
        "rol": "Bombardero estratégico pesado",
        "caracteristica_clave": "Longevidad increíble, gran carga útil, columna vertebral de la flota de bombarderos",
        "generacion": "Primera/Segunda (modernizado)",
        "velocidad_maxima": "1047 km/h (Mach 0.86)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "8 × Pratt & Whitney TF33-P-3/103",
        "dato_historico": "Con más de 60 años en servicio, es un icono de la aviación militar y se espera que vuele hasta la década de 2050.",
        "estado_operativo": "Operativo"
    },
    "F-111 Aardvark": {
        "rol": "Caza-bombardero de ataque interdictor",
        "caracteristica_clave": "Avión de ala de geometría variable, bombardeo a baja cota y alta velocidad",
        "generacion": "Tercera",
        "velocidad_maxima": "Mach 2.5 (aprox. 2655 km/h)",
        "techo_operativo": "60,000 ft (aprox. 18,200 m)",
        "motores": "2 × Pratt & Whitney TF30",
        "dato_historico": "Apodado 'Aardvark' por su nariz larga y su capacidad de 'olfatear' el terreno a baja altura.",
        "estado_operativo": "Retirado"
    },
    "F-14 Tomcat": {
        "rol": "Caza de superioridad aérea, interceptor naval",
        "caracteristica_clave": "Alas de geometría variable, misil AIM-54 Phoenix, icónico de la US Navy",
        "generacion": "Cuarta",
        "velocidad_maxima": "Mach 2.34 (aprox. 2485 km/h)",
        "techo_operativo": "53,000 ft (aprox. 16,150 m)",
        "motores": "2 × Pratt & Whitney TF30 (primeras versiones) o General Electric F110",
        "dato_historico": "Famoso por su aparición en la película Top Gun, era el principal interceptor de la US Navy.",
        "estado_operativo": "Retirado"
    },
    "F-4 Phantom II": {
        "rol": "Interceptor, caza-bombardero",
        "caracteristica_clave": "Versátil, biplaza, bimotor, muy producido y usado en múltiples roles",
        "generacion": "Tercera",
        "velocidad_maxima": "Mach 2.23 (aprox. 2370 km/h)",
        "techo_operativo": "60,000 ft (aprox. 18,300 m)",
        "motores": "2 × General Electric J79",
        "dato_historico": "Es uno de los aviones militares occidentales más producidos de la Guerra Fría y fue usado por muchas fuerzas aéreas.",
        "estado_operativo": "Retirado (en EE. UU., aún operativo en otros países)"
    },
    "P-51 Mustang": {
        "rol": "Caza de escolta de largo alcance",
        "caracteristica_clave": "Uno de los mejores cazas de la Segunda Guerra Mundial, motor Merlin",
        "generacion": "Primera/Segunda (de la época)",
        "velocidad_maxima": "703 km/h",
        "techo_operativo": "41,900 ft (aprox. 12,770 m)",
        "motores": "1 × Packard V-1650-7 Merlin",
        "dato_historico": "Su alcance permitió escoltar a los bombarderos aliados hasta Berlín y de regreso, cambiando el curso de la guerra aérea.",
        "estado_operativo": "Retirado"
    },
    "F-86 Sabre": {
        "rol": "Caza de superioridad aérea",
        "caracteristica_clave": "Principal caza de la Guerra de Corea, primer caza de EE. UU. con alas en flecha",
        "generacion": "Segunda",
        "velocidad_maxima": "1106 km/h (Mach 0.9)",
        "techo_operativo": "49,900 ft (aprox. 15,200 m)",
        "motores": "1 × General Electric J47",
        "dato_historico": "Combatió contra el MiG-15 soviético en la 'MiG Alley' durante la Guerra de Corea, con un alto índice de victorias.",
        "estado_operativo": "Retirado"
    },
    "B-29 Superfortress": {
        "rol": "Bombardero estratégico pesado",
        "caracteristica_clave": "Bombardero de la Segunda Guerra Mundial, el que lanzó las bombas atómicas sobre Japón",
        "generacion": "Primera",
        "velocidad_maxima": "574 km/h",
        "techo_operativo": "31,850 ft (aprox. 9,700 m)",
        "motores": "4 × Wright R-3350 Duplex-Cyclone",
        "dato_historico": "Fue el bombardero más avanzado de la Segunda Guerra Mundial y jugó un papel crucial en el fin del conflicto.",
        "estado_operativo": "Retirado"
    },
    "YF-17 Cobra": {
        "rol": "Prototipo de caza ligero",
        "caracteristica_clave": "Diseño de alta maniobrabilidad, base para el F/A-18 Hornet",
        "generacion": "Cuarta (prototipo)",
        "velocidad_maxima": "Mach 1.95 (aprox. 2070 km/h)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "2 × General Electric YJ101",
        "dato_historico": "Aunque no fue elegido como caza ligero de la USAF, su diseño evolucionó para convertirse en el exitoso F/A-18 Hornet.",
        "estado_operativo": "Prototipo"
    },
    "YF-16": {
        "rol": "Prototipo de caza ligero",
        "caracteristica_clave": "Tecnología fly-by-wire, base para el F-16 Fighting Falcon",
        "generacion": "Cuarta (prototipo)",
        "velocidad_maxima": "Mach 2.05 (aprox. 2170 km/h)",
        "techo_operativo": "55,000 ft (aprox. 16,760 m)",
        "motores": "1 × Pratt & Whitney F100",
        "dato_historico": "Ganador del concurso LWF (Lightweight Fighter) que llevó al desarrollo del F-16 Fighting Falcon.",
        "estado_operativo": "Prototipo"
    },
    "YF-22": {
        "rol": "Prototipo de caza avanzado",
        "caracteristica_clave": "Ganador del concurso ATF, precursor directo del F-22 Raptor",
        "generacion": "Quinta (prototipo)",
        "velocidad_maxima": "Mach 2.2+ (estimado)",
        "techo_operativo": "65,000 ft (aprox. 19,812 m)",
        "motores": "2 × General Electric YF120 o Pratt & Whitney YF119 (prototipos)",
        "dato_historico": "Compitió y ganó contra el YF-23 para convertirse en el F-22 Raptor, definiendo la siguiente generación de cazas.",
        "estado_operativo": "Prototipo"
    },
    "SR-71 Blackbird": {
        "rol": "Avión de reconocimiento estratégico de largo alcance y alta velocidad",
        "caracteristica_clave": "Velocidad y altitud extremas, el avión tripulado más rápido del mundo",
        "generacion": "Tercera",
        "velocidad_maxima": "Mach 3.3+ (aprox. 3540 km/h)",
        "techo_operativo": "85,000 ft (aprox. 25,900 m)",
        "motores": "2 × Pratt & Whitney J58",
        "dato_historico": "Mantuvo el récord del avión tripulado más rápido del mundo y era prácticamente invulnerable a los interceptores y misiles debido a su velocidad.",
        "estado_operativo": "Retirado"
    },
    "YF-12": {
        "rol": "Prototipo de interceptor",
        "caracteristica_clave": "Versión interceptora del A-12/SR-71, rápido y de gran altitud",
        "generacion": "Tercera",
        "velocidad_maxima": "Mach 3.2 (aprox. 3400 km/h)",
        "techo_operativo": "80,000 ft (aprox. 24,400 m)",
        "motores": "2 × Pratt & Whitney J58",
        "dato_historico": "Fue el interceptor tripulado más grande y pesado jamás construido, diseñado para destruir bombarderos soviéticos.",
        "estado_operativo": "Prototipo/Retirado"
    },
    "YF-107 Ultra Sabre": {
        "rol": "Prototipo de caza-bombardero",
        "caracteristica_clave": "Toma de aire dorsal distintiva, alta velocidad",
        "generacion": "Segunda",
        "velocidad_maxima": "Mach 2+ (estimado)",
        "techo_operativo": "50,000 ft+ (estimado)",
        "motores": "1 × Pratt & Whitney J75",
        "dato_historico": "Perdió el contrato de producción ante el F-105 Thunderchief, a pesar de sus impresionantes capacidades.",
        "estado_operativo": "Prototipo"
    },
    "YF-118G Bird of Prey": {
        "rol": "Prototipo de avión furtivo experimental",
        "caracteristica_clave": "Tecnología furtiva avanzada, diseño sin cola y fuselaje plano",
        "generacion": "Cuarta (experimental)",
        "velocidad_maxima": "Subsonico",
        "techo_operativo": "20,000 ft (aprox. 6,100 m)",
        "motores": "1 × Pratt & Whitney JT15D-5C",
        "dato_historico": "Fue un proyecto secreto de demostración tecnológica de sigilo de Lockheed Martin que sentó las bases para futuros diseños furtivos.",
        "estado_operativo": "Prototipo/Experimental"
    },
    "C-17 Globemaster III": {
        "rol": "Transporte aéreo estratégico y táctico",
        "caracteristica_clave": "Gran capacidad de carga, puede operar en pistas cortas y no preparadas",
        "generacion": "N/A (transporte)",
        "velocidad_maxima": "830 km/h (Mach 0.65)",
        "techo_operativo": "45,000 ft (aprox. 13,716 m)",
        "motores": "4 × Pratt & Whitney F117-PW-100",
        "dato_historico": "La espina dorsal de la capacidad de transporte aéreo de largo alcance de la USAF, capaz de transportar tanques y vehículos pesados.",
        "estado_operativo": "Operativo"
    },
    "C-130 Hercules": {
        "rol": "Transporte aéreo táctico",
        "caracteristica_clave": "Extremadamente versátil, robusto, puede operar en pistas cortas y sin pavimentar",
        "generacion": "N/A (transporte)",
        "velocidad_maxima": "592 km/h",
        "techo_operativo": "33,000 ft (aprox. 10,060 m)",
        "motores": "4 × Allison T56-A-15 (turbohélice)",
        "dato_historico": "Uno de los aviones militares más longevos y versátiles de la historia, en producción continua desde 1954.",
        "estado_operativo": "Operativo"
    },
    "KC-135 Stratotanker": {
        "rol": "Reabastecimiento en vuelo y transporte",
        "caracteristica_clave": "Avión cisterna icónico, gran alcance y capacidad de combustible",
        "generacion": "N/A (cisterna)",
        "velocidad_maxima": "933 km/h (Mach 0.90)",
        "techo_operativo": "50,000 ft (aprox. 15,240 m)",
        "motores": "4 × CFM International F108-CF-100 (turbofan)",
        "dato_historico": "Ha sido la columna vertebral de la fuerza de reabastecimiento en vuelo de la USAF durante décadas, extendiendo el alcance de todas las ramas.",
        "estado_operativo": "Operativo"
    },
    "KC-46 Pegasus": {
        "rol": "Reabastecimiento en vuelo y transporte",
        "caracteristica_clave": "Avión cisterna moderno basado en el Boeing 767, con capacidades avanzadas",
        "generacion": "N/A (cisterna)",
        "velocidad_maxima": "914 km/h (Mach 0.86)",
        "techo_operativo": "40,100 ft (aprox. 12,200 m)",
        "motores": "2 × Pratt & Whitney PW4062",
        "dato_historico": "Diseñado para reemplazar a la envejecida flota de KC-135, proporcionando capacidades de reabastecimiento y transporte de nueva generación.",
        "estado_operativo": "Operativo"
    },
    "E-3 Sentry AWACS": {
        "rol": "Alerta Temprana y Control Aerotransportado (AEW&C)",
        "caracteristica_clave": "Radar distintivo en cúpula dorsal, comando y control en el aire",
        "generacion": "N/A (vigilancia)",
        "velocidad_maxima": "855 km/h (Mach 0.77)",
        "techo_operativo": "41,000 ft (aprox. 12,500 m)",
        "motores": "4 × Pratt & Whitney TF33-PW-100A (turbofan)",
        "dato_historico": "Reconocible por su cúpula de radar rotatoria, es esencial para la gestión del espacio aéreo y la coordinación de operaciones de combate.",
        "estado_operativo": "Operativo"
    },
    "RQ-4 Global Hawk": {
        "rol": "Vehículo Aéreo No Tripulado (UAV) de vigilancia estratégica",
        "caracteristica_clave": "Gran altitud, larga resistencia, reconocimiento ISR de alta resolución",
        "generacion": "N/A (UAV)",
        "velocidad_maxima": "629 km/h",
        "techo_operativo": "60,000 ft (aprox. 18,290 m)",
        "motores": "1 × Rolls-Royce F137 (turbofan)",
        "dato_historico": "Capaz de volar a gran altitud durante más de 30 horas, proporcionando inteligencia y reconocimiento sobre vastas áreas.",
        "estado_operativo": "Operativo"
    },
    "AC-130 Specter/Ghost": {
        "rol": "Cañonero (Gunship) de apoyo aéreo cercano",
        "caracteristica_clave": "Variante de ataque del C-130, armado con cañones pesados para apoyo terrestre",
        "generacion": "N/A (gunship)",
        "velocidad_maxima": "480 km/h",
        "techo_operativo": "25,000 ft (aprox. 7,620 m)",
        "motores": "4 × Allison T56-A-15 (turbohélice)",
        "dato_historico": "Famoso por su letal capacidad de fuego en tierra, operando en misiones de apoyo a fuerzas especiales y protección de convoyes.",
        "estado_operativo": "Operativo"
    }
}

def get_jet_info(jet_name):
    """Devuelve información sobre un avión de combate si está en la base de conocimiento, manejando búsquedas parciales."""
    normalized_query = jet_name.lower().strip()

    # Buscar coincidencias exactas primero (insensible a mayúsculas/minúsculas y espacios)
    exact_match = None
    for key in fighter_jets:
        if key.lower() == normalized_query:
            exact_match = key
            break

    if exact_match:
        jet = fighter_jets.get(exact_match)
        print(f"Información sobre el {exact_match}:")
        for key, value in jet.items():
            print(f"  {key.replace('_', ' ').capitalize()}: {value}")
        return

    # Si no hay coincidencia exacta, buscar coincidencias parciales
    partial_matches = []
    for key, info in fighter_jets.items():
        if normalized_query in key.lower():
            partial_matches.append(key)

    if len(partial_matches) == 1:
        # Si solo hay una coincidencia parcial, mostrar su información
        jet = fighter_jets.get(partial_matches[0])
        print(f"Información sobre el {partial_matches[0]}:")
        for key, value in jet.items():
            print(f"  {key.replace('_', ' ').capitalize()}: {value}")
    elif len(partial_matches) > 1:
        # Si hay múltiples coincidencias parciales, pedir aclaración
        print(f"El término '{jet_name}' coincide con varios aviones. ¿A cuál te refieres?")
        for match in partial_matches:
            print(f"- {match}")
        print("Por favor, sé más específico.")
    else:
        # Si no hay ninguna coincidencia
        print(f"Lo siento, no tengo información específica sobre el '{jet_name}' en esta base de datos simple.")
        print("Aviones disponibles: ", ", ".join(fighter_jets.keys()))

# Ejemplo de uso:
print("\n--- Ejemplos de consultas ---")
get_jet_info("F-22 Raptor")
get_jet_info("F-16 Fighting Falcon")
get_jet_info("F-4 Phantom")
```
